#### Q5 (b) (i)

```python
k_list = [(1,10),(1,50),(1,100),(1,200),(-100,100)]

def a_k_1(k):
    return (-1 / (6 * np.pi * k)) * (np.sin(np.pi * k / 2) + np.exp(-1j * np.pi / 2 * (k + 1)) * (np.exp(-1j * np.pi * k) - 1))

a_0_1 = 0.75
T_0_1 = 20

f_0_1 = 1 / T_0_1
t_1 = np.arange(-25, 25, 0.01)

fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(12, 12))

for index, k_tuple in enumerate(k_list):
    k_min = k_tuple[0]
    k_max = k_tuple[1]
    x_t_1 = a_0_1
    for k in range(k_min, k_max + 1):
        if not k == 0:
            x_t_1 += a_k_1(k) * np.exp(1j * 2 * np.pi * k * f_0_1 * t_1)
    axes_index = (index // 2, index % 2)
    axes[axes_index].plot(t_1, x_t_1)
    title = "k = " + str(k_min) + " .. " + str(k_max)
    axes[axes_index].set_title(title)
    
axes[2, 1].remove()

fig.tight_layout()
plt.show()
```

#### Q5 (b) (ii)

```python
def a_k_2(k):
    return (4j * np.pi * k - 3 * np.exp(1j * 4/3 * np.pi * k) + 3)/(4 * np.pi ** 2 * k ** 2) - (6j * np.pi * k + 9 * np.exp(-1j * 2/3 * np.pi * k) - 9)/(6 * np.pi ** 2 * k ** 2)

a_0_2 = 1
T_0_2 = 3

f_0_2 = 1 / T_0_2
t_2 = np.arange(-5, 5, 0.01)

fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(12, 12))

for index, k_tuple in enumerate(k_list):
    k_min = k_tuple[0]
    k_max = k_tuple[1]
    x_t_2 = a_0_2
    for k in range(k_min, k_max + 1):
        if not k == 0:
            x_t_2 += a_k_2(k) * np.exp(1j * 2 * np.pi * k * f_0_2 * t_2)
    axes_index = (index // 2, index % 2)
    axes[axes_index].plot(t_2, x_t_2)
    title = "k = " + str(k_min) + " .. " + str(k_max)
    axes[axes_index].set_title(title)
    
axes[2, 1].remove()

fig.tight_layout()
plt.show()
```

#### Q5 (b) (iii)

```python
def a_k_3(k):
    return 0.25 * (np.exp(-1j * np.pi * k)/(k**2 - 1) * (0.32 * (np.exp(1j * np.pi * k) + 1)) + (2 / k) * (0.64 * 1j * (np.exp(-1j * np.pi * k) - 1)))

a_0_3 = 1
T_0_3 = 3

f_0_3 = 1 / T_0_3
t_3 = np.arange(-5, 5, 0.01)

fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(12, 12))

for index, k_tuple in enumerate(k_list):
    k_min = k_tuple[0]
    k_max = k_tuple[1]
    x_t_3 = a_0_3
    for k in range(k_min, k_max + 1):
        if k != 0 and k * k != 1:
            x_t_3 += a_k_3(k) * np.exp(1j * 2 * np.pi * k * f_0_3 * t_3)
    axes_index = (index // 2, index % 2)
    axes[axes_index].plot(t_3, x_t_3)
    title = "k = " + str(k_min) + " .. " + str(k_max)
    axes[axes_index].set_title(title)
    
axes[2, 1].remove()

fig.tight_layout()
plt.show()
```

#### Q5 (b) (iv)

```python
def a_k_4(k):
    return 0.5 * ((1 - np.exp(-1j * np.pi * k - 1)) / (1j * np.pi * k + 1))

a_0_4 = 0.32
T_0_4 = 2

f_0_4 = 1 / T_0_4
t_4 = np.arange(-5, 5, 0.01)

fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(12, 12))

for index, k_tuple in enumerate(k_list):
    k_min = k_tuple[0]
    k_max = k_tuple[1]
    x_t_4 = a_0_4
    for k in range(k_min, k_max + 1):
        if k != 0:
            x_t_4 += a_k_4(k) * np.exp(1j * 2 * np.pi * k * f_0_4 * t_4)
    axes_index = (index // 2, index % 2)
    axes[axes_index].plot(t_4, x_t_4)
    title = "k = " + str(k_min) + " .. " + str(k_max)
    axes[axes_index].set_title(title)
    
axes[2, 1].remove()

fig.tight_layout()
plt.show()
```
